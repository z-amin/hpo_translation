```python
import matplotlib.pyplot as plt
import numpy as np


def plot_triple_graph(models: list[str], metrics: list[str], scores: dict[str, list[float]]):
    bar_width = 0.2
    index = np.arange(len(models))

    plt.bar(index - bar_width, scores[models[0]], width=bar_width, label=metrics[0])
    plt.bar(index, scores[models[1]], width=bar_width, label=metrics[1])
    plt.bar(index + bar_width, scores[models[2]], width=bar_width, label=metrics[2])

    plt.xlabel('Models')
    plt.ylabel('Scores')
    plt.title('Evaluation metrics for pre-trained models')
    plt.xticks(index, models)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
metrics = ['SacreBLEU', '`WER``', 'Semantic similarity']
models = ['Helsinki NLP', 'Meta NLLB', 'Google Madlad']
scores = {
    'Helsinki NLP': [42.8, 49.5, 90.9],
    'Meta NLLB': [39.0, 44.9, 90.1],
    'Google Madlad': [0.90, 0.80, 0.85]
}

plot_triple_graph(models, metrics, scores)
```
